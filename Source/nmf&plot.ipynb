```python
import numpy as np
import matplotlib.pyplot as plt

def calcula_custo(A, W, H):
    return 0.5 * np.sum((A - np.dot(W, H)) ** 2)

def nmf(A, k, maxit=1000):
    m, n = A.shape
    W = np.random.rand(m, k) * 0.0009  
    H = np.random.rand(k, n) * 0.0009  
    custos = []

    for i in range(maxit):
        H = H * (np.dot(W.T, A) / (np.dot(np.dot(W.T, W), H) + 1e-9))
        W = W * (np.dot(A, H.T) / (np.dot(W, np.dot(H, H.T)) + 1e-9))
        
        custo = calcula_custo(A, W, H)
        custos.append(custo)

    return W, H, custos

np.random.seed(0)

A = np.random.rand(10, 10)

k = 3
maxit = 1000

W, H, custos = nmf(A, k, maxit)

plt.plot(range(maxit), custos)
plt.xlabel('Iterações')
plt.ylabel('Função de Custo')
plt.title('Convergência da Função de Custo no NMF')
plt.show()
```
